```python
n = 6
for i in range(1, n + 1):
    print(" " * (n - i), end="")

    # Left side letters A to current
    for j in range(1, i + 1):
        print(chr(64 + j), end="")

    # Right side letters from current - 1 back to A
    for j in range(i - 1, 0, -1):
        print(chr(64 + j), end="")

    print()
```
