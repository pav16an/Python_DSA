```python
#sliding window
string = 'hello world'
sub = 'world'
for i in range(len(string)):
    result = ''
    for j in range(i,i+len(sub)):
        if j < len(string):
            result+=string[j]
    if result == sub:
        print(i)
        break
    
```

```python
user = input()
print(user)
if len(user) < 8:
    print('password size must be greater than 8')
else:
    has_islower = False
    has_isupper = False
    has_isdigit = False
    has_isspecial = False
    
    for ch in user:
        if ch.islower:
            has_islower = True
        if ch.isupper:
            has_isupper = True
        if ch.isdigit:
            has_isdigit = True
        if ch in '!@#$%^&*()-+[]{}<>?~':
            has_isspecial = True
    
    if all([has_islower, has_isupper, has_isdigit, has_isspecial]):
        print("✅ Password is strong")
    else:
        if not has_islower:
            print("❌ At least one lowercase letter is required")
        if not has_isupper:
            print("❌ At least one uppercase letter is required")
        if not has_isdigit:
            print("❌ At least one numeric value is required")
        if not has_isspecial:
            print("❌ At least one special character is required")


```

```python
#reverse
n = 1234
val = 0
while n > 0:
    reminder = n % 10
    n = n // 10
    val = (val*10) + reminder
print(val)
```

```python
#palindrome
n = 1232
dup = n
rev = 0

while dup > 0:
        remind = dup%10
        dup = dup//10
        rev = (rev*10) + remind
print(rev)

if rev == n:
    print("palindrome")
else:
    print("Not Palindrome")
```

```python
#armstrong
n = 1634
dup = n
res = 0
power = len(str(n))
while dup > 0:
    remind = dup%10
    dup = dup//10
    res = (remind**power) + res
print(res)
if res == n:
    print("armstrong")
else:
    print("Not armstrong")
```
