```python
def recur(num):
    if num<0:
        return num
    return num+recur(num-1)

recur(5)
```

```python
#reverse array
n = int(input())
arr = []
for i in range(n):
    arr.append(int(input()))
l = arr[0]
r = arr[-1]
def reverse(arr,l,r):
    if l >= r:
        return 
    l = r
    r = l
    return reverse(arr,l+1,r-1)
```
