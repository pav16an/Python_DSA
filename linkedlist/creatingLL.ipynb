```python
class Node :
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
    
    def setdata(self,data):
        self.data = data

    def getdata(self):
        return self.data
    
    def setnext(self,next):
        self.next = next

    def getnext(self):
        return self.next
    

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)


head.setnext(node1)
node1.setnext(node2)
node2.setnext(node3)
node3.setnext(None)


def Traversal(head):
    temp = head
    while(temp):
        print(temp.getdata(),end="->")
        temp = temp.getnext()
    print("None")

Traversal(head)
```

```python
#finding length of ll
def length(head):
    len = 0
    while head:
        len+=1
        head = head.getnext()
    return len

length(head)
```

```python
#finding length using recursion
def lenght_rec(head):
    if head is None:
        return 0
    return 1 + lenght_rec(head.getnext())

lenght_rec(head)
```

```python

def insert_at_pos(head,data,k):
    new_node = Node(data)
    if (k>length(head) or k<0 ):
        print("Invalid position")
        return head
    if k == 0:
        new_node.setnext(head)
        head = new_node
    else:
        temp = head
        i = 0
        while(i < k-1):
            temp = temp.getnext()
            i += 1
        new_node.setnext(temp.getnext())
        temp.setnext(new_node)
    return head

```

```python
def find_circle(head):
    if head is None:
        return None
    slow = head
    fast = head
    while fast and fast.getnext():
        slow = slow.getnext()
        fast = fast.getnext().getnext()
        if fast and slow.getdata() == fast.getdata():
            print("Linked list have a cycle",slow.getdata())
            break
    else:
        print("linked list don't have a cycle")

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
head.setnext(node1)
node1.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(node2)  # Creating a cycle for testing


find_circle(head)
```

```python
def find_pos_of_cycle(head):
    if head is None:
        return None
    iscyclePresent = False
    fast = head
    slow = head
    while fast and fast.getnext():
        slow = slow.getnext()
        fast = fast.getnext().getnext()
        if slow.getdata() == fast.getdata():
            iscyclePresent = True
            break
    if not iscyclePresent:
        return None
    temp = head
    while slow.getdata() != temp.getdata():
        slow = slow.getnext()
        temp = temp.getnext()

    return temp.getdata()


head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
head.setnext(node1)
node1.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(node3)

find_pos_of_cycle(head)
```

```python
#Reversing a linked list
def reverse_linked_list(head):
    prev = None
    current = head
    while current:
        next_node = current.getnext()  # Store next node
        current.setnext(prev)           # Reverse the link
        prev = current                  # Move prev to current
        current = next_node             # Move to next node
    return prev  # New head of the reversed list

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)
head.setnext(node1)
node1.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(None)

head = reverse_linked_list(head)
Traversal(head)
```

```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def setdata(self, data):
        self.data = data

    def getdata(self):
        return self.data

    def setnext(self, next):
        self.next = next

    def getnext(self):
        return self.next


# Linked List 1: 1 -> 2 -> 7 -> 6
head = Node(1)
node1 = Node(2)
node2 = Node(7)
node3 = Node(6)

head.setnext(node1)
node1.setnext(node2)
node2.setnext(node3)

# Linked List 2: 5 -> 4 -> 3
head2 = Node(5)
node5 = Node(4)
node6 = Node(3)

head2.setnext(node5)
node5.setnext(node6)
node6.setnext(None)


# Print function
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.getdata(), end="->")
        current = current.getnext()
    print("None")


# Bubble Sort for Linked List
def sort_linked_list(head):
    if head is None:
        return head

    swapped = True
    while swapped:
        swapped = False
        current = head
        while current.getnext() is not None:
            next_node = current.getnext()
            if current.getdata() > next_node.getdata():
                # Swap data
                temp = current.getdata()
                current.setdata(next_node.getdata())
                next_node.setdata(temp)
                swapped = True
            current = current.getnext()


# Printing original lists
print("Original List 1:")
print_linked_list(head)

print("Original List 2:")
print_linked_list(head2)

# Sorting both lists
sort_linked_list(head)
sort_linked_list(head2)

# Printing after sorting
print("Sorted List 1:")
print_linked_list(head)

print("Sorted List 2:")
print_linked_list(head2)
```
